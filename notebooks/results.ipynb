```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/home/chri6578/Documents/CES/logs/ATD.log', header=None)
column_data = data[0] 
plt.figure(figsize=(4, 3), dpi=240)
sns.set(style="whitegrid")
sns.histplot(column_data, kde=True, stat="density", bins=30, color="skyblue", edgecolor="None")
plt.xlabel("ATD($\mathcal{S}, \mathcal{S}\'$)", fontsize=14)
plt.ylabel("", fontsize=14)
plt.xticks([500, 800], fontsize=12)
plt.yticks(fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("/home/chri6578/Documents/CES/results/ATD.pdf", dpi=300)
```

```python
data2 = pd.read_csv('/home/chri6578/Documents/CES/logs/ACD.log', header=None, names=['v', 'acd'])

grouped = data2.groupby('v')['acd'].agg(['mean', 'std']).reset_index()

plt.figure(figsize=(4, 3), dpi=240)

plt.plot(grouped['v'], grouped['mean'], marker='.', linestyle='-', color='b', linewidth=2, markersize=6)

# plt.errorbar(grouped['v'], grouped['mean'], yerr=grouped['std'], fmt='-', color='b', ecolor='gray', capsize=2, capthick=1)

plt.xlabel("$\log_{10}(\\bar{\\tau})$", fontsize=14)
plt.ylabel("ACD($\mathcal{S}, \mathcal{S}\', \\bar{\\tau}$)", fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.legend(['mean'])
plt.savefig("/home/chri6578/Documents/CES/results/ACD.pdf", dpi=300)
```

```python
data3 = pd.read_csv('/home/chri6578/Documents/CES/logs/OD.log', header=None, names=['k', 'od'])

grouped3 = data3.groupby('k')['od'].agg(['mean', 'std']).reset_index()

plt.figure(figsize=(3, 3), dpi=240)


plt.bar(grouped3['k'], grouped3['mean'], yerr=grouped3['std'], color='skyblue', edgecolor='None', capsize=5)

plt.xlabel("$k$", fontsize=14)
plt.ylabel("OD($\mathcal{S}, \mathcal{S}\', k$)", fontsize=14)
plt.xticks([2,3,4,5],fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.ylim([0, 2])
plt.legend(['mean'])
plt.savefig("/home/chri6578/Documents/CES/results/OD.pdf", dpi=300)
```
